```python
from pylab import *
x = linspace(0,10*pi,9999)
def g(x):
    return cos(x)**2 / sqrt(maximum(1,2*x-1))
title('cos(x)^2 / sqrt(max(1,2x-1)',color="red")
plot(x,g(x),color="gray")
show()
```

```python
from pylab import *
a,b,n = 0,50,10000
h = 0.0001
x = linspace(a,b,n)
def f(x):
    return (2+sin(x/pi))/(2-sin(x/pi))
def df(x,h):
    return (f(x+h)-f(x))/h
plot(x,f(x))
plot(x,df(x,h))
xlabel("X")
ylabel("Y")
show()
```
